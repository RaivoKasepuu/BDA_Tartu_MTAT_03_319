### Clean the dataset, according to the rules below. Write down, which columns and rows were affected and how many rows are left after the cleaning? Please, provide reasoning for your decision. The rest of the homework depends on these results, so be careful with this task.

```python
print('General data cleaning rule # 1. It is not good practice to have spaces in column names. Lets replace spaces with "_": ')
df.columns = [c.replace(' ', '_') for c in df.columns]

print('General data cleaning rule " 2. Lets avoid as much as possible deleting rows/columns from dataFrame. Each deleting must be well explained and with clear reason')
```

```python

print('Rule #1 "App feature must contain unique values" implementation:')
print('"App" column unique values are ' + str(df.App.unique()))
print()
print("Dataframe has " + str(df.shape[0]) + " rows" )
print(df[df.App.duplicated()])
print("Duplicates removed. Dataframe has now " + str(df.shape[0]) + " rows" )
```

```python

print('Rule #2 "Category feature must contain 15 different values" implementation:')
print()
print('"Category" column unique values are ' + str(df.Category.unique()))
df.loc[df.Category.isin(['SHOPPING', 'TRAVEL_AND_LOCAL', 'DATING', 'BOOKS_AND_REFERENCE', 'VIDEO_PLAYERS', 'EDUCATION', 'ENTERTAINMENT', 'MAPS_AND_NAVIGATION', 'FOOD_AND_DRINK', 'HOUSE_AND_HOME', 'AUTO_AND_VEHICLES', 'LIBRARIES_AND_DEMO', 'WEATHER', 'ART_AND_DESIGN', 'EVENTS', 'COMICS', 'PARENTING', 'BEAUTY', '1.9']), 'Category'] = 'OTHER'
df.Category.value_counts()
```

```python

print('Rule #3 "Rating must be positive and not exceed 5" implementation:')
# Let's have a look first: 
print('"Ratings" column unique values are ' + str(df.Rating.unique()))
# if Rating > 5, it will be replaced by 5. 
df.loc[df["Rating"] > 5, "Rating"] = 5
# nan will be replaced with Rating mean value, all values rounded
df['Rating'] = df['Rating'].fillna(df.Rating.mean()).round(1)
# Let's check again: 
print('"Ratings" column new unique values are: ' + str(df.Rating.unique()))
print('"Ratings" column max value is: ' + str(df.Rating.max()))
print('"Ratings" column min value is: ' + str(df.Rating.min()))
print(df.Rating)
```

```python

print('Rule #4 "Amount of reviews must be positive(non-negative)" implementation:')
# Let's have a look first: 
print('"Reviews" column unique values are ' + str(df.Reviews.unique()))
# Lets convert string to numeric:
df["Reviews"] = pd.to_numeric(df["Reviews"].str.replace('[^.0-9]', ''))
print(df.Reviews)
print("min value is: " + str(df.Reviews.min()))
```

```python
# TEHTUD - Raivo
print('Rule #5 "Size must be positive" implementation:')
# Let's have a look first: 
print('"Size" column unique values are ' + str(df.Size.unique()))

# Let's fix k and M problem with function:
def kilomega_to_float(x):
   
    if type(x) == float or type(x) == int:
      return x

    if 'k' in x:
      return float(x.replace('k', '')) * 1000
    
    if 'M' in x:
      return float(x.replace('M', '')) * 1000000
    return 0.0

df['Size'] = df['Size'].apply(kilomega_to_float)
```

```python
# check
print(df.Size)
```

```python

print('Rule #6 "Amount of installs must be positive" implementation:')
# Let's have a look first: 
print('"Installs" column unique values are: ' + str(df.Installs.unique()))
# Let's replace "Free" with "0"
df.loc[df["Installs"] == "Free", "Installs"] = "0"
# Let's remove "+" and "," and convert object to numeric
df['Installs'] = df['Installs'].str.replace('+','')
df['Installs'] = df['Installs'].str.replace(',','')
df["Installs"] = pd.to_numeric(df["Installs"])

print('"Installs" column new unique values are: ' + str(df.Installs.unique()))
print(df.Installs)
```

```python
#check
print(df.Installs)
print(df.Installs.min())
print(df.Installs.max())

```

```python

print('Rule #7 "Type of payment should be either “Free” or “Paid”" implementation:')
# Let's have a look first: 
print('"Type" column unique values are ' + str(df.Type.unique()))
# Let's replace values "0" with "Free"
df['Type'] = df['Type'].str.replace('0','Free')

# Lets examine nan rows. Do we have "Price" information for decision? 
# row indexes without Type values (nan)
print(df[df['Type'].isna().tolist()])
print(df[df['Type'].isna().tolist()].index.tolist()[0])

# lets check Price on those rows: 
print(df.loc[[df[df['Type'].isna().tolist()].index.tolist()[0]], ["Price"]])


# As Price is 0 on row 9148, nan will be filled with "Free"
df['Type'] = df['Type'].fillna("Free")
print('"Type" column new unique values are ' + str(df.Type.unique()))
```

```python
# check
print('"Type" column new unique values are ' + str(df.Type.unique()))
```

```python

print('Rule #8 "Price has to be of type float64" implementation:')
# Let's have a look on Price unique values: 
print('"Price" column unique values are ' + str(df.Price.unique()))
#Let's convert all non numeric values to numeric
df["Price"] = pd.to_numeric(df["Price"].str.replace('[^.0-9]', ''))
# Let's check now:
print('"Price" column unique values are ' + str(df.Price.unique()))
print(df.Price)
```

```python

print('Rule #9 "NA values should be replced where it is possible" implementation:')
df_rows_with_nan = df[df.isna().any(axis=1)]
print('Rows with "nan" value: ')
df_rows_with_nan.head(15)
print("As nan data is only in columns Current_Ver and Android_Ver, there is no good idea to replace this with meaningful data.")
print("There is no need to delete those rows information due missing version information. Data in other columns on those rows has higher value. ")
```
